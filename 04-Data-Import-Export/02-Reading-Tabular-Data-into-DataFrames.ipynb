# Anscombe's quartet

Anscombe's quartet comprises four datasets that have nearly identical simple descriptive statistics, yet appear very different when graphed. Each dataset consists of eleven (x,y) points. They were constructed in 1973 by the statistician Francis Anscombe to demonstrate both **the importance of graphing data before analyzing it** and **the effect of outliers on statistical properties**. He described the article as being intended to counter the impression among statisticians that "numerical calculations are exact, but graphs are rough".

## `read_json()`

`anscombe.json` contains a copy of [Anscombe's quartet](https://en.wikipedia.org/wiki/Anscombe%27s_quartet). The file contains 11 data points for 4 series. These series can be described by a similar statistics but in fact are very different from each other when plotted:

```python
anscombe = pd.read_json("sample_data/anscombe.json")
anscombe.head(20)
```

## Visualizing with Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import seaborn as sns

# Show the results of a linear regression within each dataset
sns.lmplot(x="X", y="Y", col="Series", hue="Series", data=anscombe,
           col_wrap=2, ci=None, palette="muted", height=4);
```

## Filter rows

```python
anscombe[anscombe.Series == 'I']
```

```python
anscombe[anscombe.X >= 14]
```

```python
# Splitting the DataFrame into 4 based on their Series values
anscombe_I = anscombe[anscombe.Series == 'I']
anscombe_II = anscombe[anscombe.Series == 'II']
anscombe_III = anscombe[anscombe.Series == 'III']
anscombe_IV = anscombe[anscombe.Series == 'IV']
```

Using `describe()` method to get the statistics:

```python
# Applying the descibe method on all the 4 DataFrames
desc_I = anscombe_I.describe()
desc_II = anscombe_II.describe()
desc_III = anscombe_III.describe()
desc_IV = anscombe_IV.describe()

# Renaming the column names
desc_I.columns = ['X_I', 'Y_I']
desc_II.columns = ['X_II', 'Y_II']
desc_III.columns = ['X_III', 'Y_III']
desc_IV.columns = ['X_IV', 'Y_IV']

# Join all the summaries together
summary = desc_I.join([desc_II, desc_III, desc_IV])

# Reordering the columns
summary[['X_I', 'X_II', 'X_III', 'X_IV', 'Y_I', 'Y_II', 'Y_III', 'Y_IV']]
```

Let's create a new Excel file using `anscombe` and `ca_housing` DataFrames that we imported above.

```python
# First create a writer object with a filename
writer = pd.ExcelWriter('sample_data/pandas-example.xlsx')

# Add DataFrames that need to be exported to the sheet defined above
anscombe.to_excel(writer, sheet_name='anscombe', index=False)
ca_housing.to_excel(writer, sheet_name='ca_housing', index=False)

# Write it to disk
writer.close()
```
