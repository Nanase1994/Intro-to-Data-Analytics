2. Fill the null values with a string to represent missing values (e.g., "unspecified").
3. Estimate the values using an algorithm such as ML.

##### User rating distribution: The histogram shows where most users’ average ratings are concentrated. If the distribution skews toward higher scores, it suggests that users generally give positive evaluations. If the ratings are more spread out or cluster at lower values, it indicates more diverse or negative user experiences.

##### Mean vs. Median comparison: If the mean is higher than the median, it implies that a small group of users gave particularly high ratings, pulling the average up. Conversely, if the mean is lower, some users gave very low ratings, dragging the average down. This helps identify whether the distribution is skewed.

##### User satisfaction trend: The further the distribution shifts to the right (higher ratings), the higher the overall user satisfaction. A noticeable concentration on the left side (lower ratings) may indicate issues with product quality or user experience.

```python
import matplotlib.pyplot as plt
import seaborn as sns
# "Distribution of Users’ Average Ratings and Satisfaction Tendencies"
user_avg_rating = merged_df.groupby("author_id")["review_rating"].mean()

sns.set_style("whitegrid")

plt.figure(figsize=(10,6))

sns.histplot(user_avg_rating, bins=20, kde=True, color="skyblue", edgecolor="black", alpha=0.7)

mean_val = user_avg_rating.mean()
median_val = user_avg_rating.median()

plt.axvline(mean_val, color="red", linestyle="--", linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.axvline(median_val, color="green", linestyle="-.", linewidth=2, label=f"Median = {median_val:.2f}")

plt.title("Distribution of Users’ Average Ratings", fontsize=16, fontweight="bold")
plt.xlabel("Average Rating per User", fontsize=14)
plt.ylabel("Number of Users", fontsize=14)

sns.despine()

plt.legend()

plt.show()
```
